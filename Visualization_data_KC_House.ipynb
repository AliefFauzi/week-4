```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from ipywidgets import interact,IntSlider
from google.colab import drive
drive.mount('/content/drive')
```

```python
file_path='/content/drive/MyDrive/kc_house_data.csv'
datas=pd.read_csv(file_path)
datas.head()
```

```python
unique_value_counts=datas.nunique()
print(unique_value_counts)
```

```python
summary=datas.describe()
summary
```

```python
for column in datas.columns:
  if datas[column].dtype =='float64':
    plt.figure(figsize=(8,6))
    sns.histplot(datas[column],kde=True,color='skyblue')
    plt.title(f'Distrubusi {column}', size=15)
    plt.xlabel(column,size=12)
    plt.ylabel('Frequency',size=12)
    plt.show()
```

```python
for column in datas.columns:
  if datas[column].dtype=='float64':
    plt.figure(figsize=(8,6))
    sns.histplot(datas[column],kde=True,color='khaki')
    plt.title(f'Distribusi {column}',size=15)
    plt.xlabel(column,size=12)
    plt.ylabel('frequency',size=12)
    plt.show()
```

```python
#jenis plot dan ukuran nya
fig, axes = plt.subplots(3, 2, figsize=(15, 15))

columns_to_process = ['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors']


if len(columns_to_process) != 6:
    raise ValueError("Jumlah kolom yang ingin Anda proses harus sama dengan jumlah subplot (6).")

for i, column in enumerate(columns_to_process):
    sns.histplot(datas[column], kde=True, color='khaki', ax=axes[i//2, i%2])
    axes[i//2, i%2].set_title(f'Distribusi {column}', size=12)
    axes[i//2, i%2].set_xlabel(column, size=10)
    axes[i//2, i%2].set_ylabel('Frequency', size=10)

plt.tight_layout()  # Mengatur tata letak subplot agar tidak tumpang tindih
plt.show()
```

```python
def plot_distribution(feature, bin_width):
    plt.figure(figsize=(8, 6))
    sns.histplot(datas[feature], bins=bin_width, kde=True, color='khaki')
    plt.title(f'Distribusi {feature} (Bin Width: {bin_width})', size=15)
    plt.xlabel(feature, size=12)
    plt.ylabel('Frequency', size=12)
    plt.show()

# Interaksi untuk memilih fitur dan bin width
interact(plot_distribution, feature=datas.columns, bin_width=IntSlider(min=1, max=50, step=1, value=10))
```

```python
#Menampilkan Boxplot
for column in datas.columns :
  if datas[column].dtype=='float64':
    plt.figure(figsize=(8,6))
    sns.boxplot(y=datas[column],color='khaki')
    plt.title(f'Boxplot of {column}',size=15)
    plt.ylabel(column,size=12)
    plt.show()
```

```python
fig, axes = plt.subplots(3, 2, figsize=(15, 15))

columns_in_fair = ['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors']


if len(columns_in_fair) != 6:
    raise ValueError("Jumlah kolom yang ingin Anda proses harus sama dengan jumlah subplot (6).")

for i, column in enumerate(columns_in_fair):
    sns.boxplot(x=datas[column], color='purple', orient='h', ax=axes[i//2, i%2])
    axes[i//2, i%2].set_title(f'Distribusi {column}', size=12)
    axes[i//2, i%2].set_xlabel(column, size=10)
    axes[i//2, i%2].set_ylabel('Frequency', size=10)

plt.tight_layout()  # Mengatur tata letak subplot agar tidak tumpang tindih
plt.show()
```

```python
sns.set(style="ticks")
sns.pairplot(datas, kind="scatter", diag_kind="kde", palette="husl")
plt.show()
```

```python
#Menampilkan Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(datas.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Heatmap Korelasi Fitur', size=15)
plt.show()
```

```python
column_in_pro=['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors']
plt.figure(figsize=(12, 8))

# Membuat violin chart untuk setiap fitur
for i, column in enumerate(column_in_pro):
    if datas[column].dtype == 'float64':
        plt.subplot(3, 3, i+1)
        sns.violinplot(y=datas[column], color='skyblue')
        plt.title(f'Violin Chart {column}', size=12)
        plt.ylabel(column, size=10)

# Menampilkan plot
plt.tight_layout()
plt.show()
```
